```python
from os import name
import warnings
warnings.simplefilter("always")

#====================================================TASK 1====================================================
print(f"\n{'#'*40} Task 1 — Data Parsing & Profile Cleaning {'#'*40}\n")
#raw student data with inconsistent formatting
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
] 
cleaned_students=[]                                             #list to store cleaned student data

#Looping through raw_students and for each student, producing a cleaned version
for student in raw_students:
  cl_name=student["name"].strip().title()                       #removing leading/trailing spaces and converting to title case
  cl_roll=int(student["roll"])                                  #converting roll number from string to integer
  cl_marks=[int(x) for x in student["marks_str"].split(",")]    #converting marks from comma-separated string to list of integers
  cleaned={"name":cl_name,"roll":cl_roll,"marks":cl_marks}      #creating a cleaned student dictionary with the cleaned name, roll number, and marks
  cleaned_students.append(cleaned)
  if(cl_roll==103):                                             #storing the cleaned name of the student with roll number 101 in both uppercase and lowercase for later display
      name103=cl_name
      upper103=name103.upper()
      lower103=name103.lower()

#For each cleaned student, verifying the name is valid containing only alphabetic characters
for cstudent in cleaned_students:
    if ("".join(cstudent["name"].split())).isalpha():           #removing spaces from the name
      validity="✓ Valid name"
    else:
      validity="✗ Invalid name"
    print("======================================")             #printing a formatted profile card for each student
    print("Student:", cstudent["name"], validity)
    print("Roll no:", cstudent["roll"])
    print("Marks:", cstudent['marks'])
    print("======================================\n")

print("UPPERCASE:",upper103)                                    #displaying the cleaned name of the student with roll number 103 in uppercase
print("lowercase:",lower103)                                    #and lowercase

#====================================================TASK 2====================================================
print(f"\n{'#'*40} Task 2 — Marks Analysis Using Loops & Conditionals {'#'*40}\n")
# Given Data
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

#Function to return Grade with the marks obtained in a subject as string parameter
def getGrade(score):
  if(score.isdigit()==False or (int(score)<0 or int(score)>100)): #check if the score is a valid integer between 0 and 100
    return "Invalid Marks"
  else:
    score=int(score)
    if(score<=100 and score>=90):
        return "A+"
    elif(score<90 and score>=80):
        return "A"
    elif(score<80 and score>=70):
        return "B"
    elif(score<70 and score>=60):
        return "C"
    elif(score<60 and score>=0):
        return "F"

#Printing the formatted report card for the student
print(f"\n{'='*30}")
print(f"Student Name: {student_name}")
print(f"{'='*30}")
print(f"{'Subject':<12} | {'Marks':>5} | {'Grade':>5}")
print(f"{'-'*30}")

for sub,mark in zip(subjects,marks):
  print(f"{sub:<12} | {mark:>5} | {getGrade(str(mark)):<5}")
  print(f"{'-'*30}")

#Statistics calculations and display

total=sum(marks)
average=round(total/len(marks),2)
hiscore= max(marks)
hisub=subjects[marks.index(hiscore)]
loscore=min(marks)
losub=subjects[marks.index(loscore)]
print(f"\n{'='*50}")
print(f"{'Total Marks:':<12} {total:>6}")
print(f"{'Average Marks:':<12} {average:>5}")
print(f"Highest Scoring Subject: {hisub:^6} \tMarks: {hiscore}")
print(f"Lowest Scoring Subject: {losub:^13} \tMarks: {loscore}")
print(f"\n{'='*50}")

#Marks Entry System for adding new subjects
newsub_count=0                                                  #Keeping track of the number of new subjects added

print(f"\nTo add a new subject, enter the subject name. Enter 'Done' as Subject Name when done.")

while True:
  new_sub=input("Enter Subject Name:").strip()
  if(new_sub.lower()=="done"):
    break
  raw_mark=input("Enter Marks:").strip()
  grade=getGrade(raw_mark)
  if(grade=="Invalid Marks"):                                   #Warning display if Invalid Marks are entered
    warnings.warn("Do not crash, and do not add invalid entries to the list.")
    continue
  new_mark=int(raw_mark)
  subjects.append(new_sub)
  marks.append(new_mark)
  newsub_count+=1
#displaying number of new subjects added and updated average
print(f"\n{'='*50}")
print(f"New Subjects Added: {newsub_count}")
print(f"Updated Average: {round(sum(marks)/len(marks),2)}")
print(f"{'='*50}")

#====================================================TASK 3====================================================
print(f"\n{'#'*40} Task 3 — Class Performance Summary {'#'*40}\n")
#Given Data
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

#initialising counters
pass_count=0    
fail_count=0
all_avgs= []

#printing the formatted class report
print(f"{'Name':<20} | {'Average':^9} | {'Status':<10}")
print(f"{'-'*41}")

for name,marks in class_data:
  avg= round(sum(marks)/len(marks),2)
  if avg>=60:
    status="Pass"
    pass_count+=1
  else:
    status="Fail"
    fail_count+=1
  all_avgs.append(avg)
  print(f"{name:<20} | {avg:^9} | {status:<10}")
print(f"\n\n")

#Computing and displaying class statistics
topper_avg=max(all_avgs)
topper_index=all_avgs.index(topper_avg)
topper_name=class_data[topper_index][0]
class_avg=round(sum(all_avgs)/len(all_avgs),2)

print(f"No. of students passed: {pass_count}")
print(f"No. of students failed: {fail_count}")
print(f"Topper Name:  {topper_name} | Topper Average:  {topper_avg}")
print(f"Class Average: {class_avg}")
#====================================================TASK 4====================================================
print(f"\n{'#'*40} Task 4 — String Manipulation Utility {'#'*40}\n")
#Given Data
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
clean_essay = essay.strip()                                         #removing leading and trailing spaces
print(clean_essay.title())                                          #converting to Title Case
print(f"No of Python occurrences: {clean_essay.count("python")}")   #counting the number of occurrences of "python" in the essay    
clean_essay=clean_essay.replace("python","Python 🐍")               # replacing "python" with "Python 🐍"
print(clean_essay)                                                  #displaying the modified essay 
sentences=clean_essay.split(". ")                                    #splitting essay into sentences at the period character
for i,sentence in enumerate(sentences,start=1):                     #printing indexed individualsentences 
  if not sentence.endswith("."):                                    #ensuring each sentence ends with a period
    sentence=sentence+"."
  print(f"{i}. {sentence}")
print(f"\n{'#'*40} End of Part-1 {'#'*40}\n")                       #displaying end of part - 1 message
  #---------------------------------END OF CODE------------------------------
```
